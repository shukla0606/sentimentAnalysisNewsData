```python
import re
import nltk
import pandas as pd
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from gensim import corpora, models

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('vader_lexicon')

def clean_article(article):
    # Remove punctuation and convert to lowercase
    article = re.sub(r'[^\w\s]', '', article.lower())
    
    # Tokenize text
    tokens = word_tokenize(article)
    
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    cleaned_article = [lemmatizer.lemmatize(word) for word in filtered_tokens]
    
    return cleaned_article

def check_mood(article):
    sid = SentimentIntensityAnalyzer()
    sentiment_score = sid.polarity_scores(article)
    
    if sentiment_score['compound'] >= 0.05:
        return 'Positive'
    elif sentiment_score['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

def find_connections(articles):
    # Tokenize and clean all articles
    cleaned_articles = [clean_article(article) for article in articles]
    
    # Create dictionary and corpus
    dictionary = corpora.Dictionary(cleaned_articles)
    corpus = [dictionary.doc2bow(article) for article in cleaned_articles]
    
    # Apply LDA
    lda_model = models.LdaModel(corpus, num_topics=3, id2word=dictionary, passes=15)
    
    # Get topics
    topics = lda_model.print_topics(num_words=5)
    
    return topics

# Extracting aspect sentiments from articles
def aspect_analysis(articles):
    aspect_sentiments = []
    for article in articles:
        # Example aspect analysis logic
        # Here, you need to implement logic to extract aspects and sentiments from each article
        # For demonstration, let's assume aspects are manually identified and sentiments are analyzed using VADER
        aspects = ["Innovation", "Cost", "Overall Plan", "Traffic Impact"]
        sentiments = ["Positive", "Negative", "Positive", "Negative"]  # Example sentiments
        
        for aspect, sentiment in zip(aspects, sentiments):
            aspect_sentiments.append({"Article": article, "Aspect": aspect, "Sentiment": sentiment})
    
    return pd.DataFrame(aspect_sentiments)

# Read news articles from Excel file
df = pd.read_excel('Assignment.xlsx')

# Extract articles from the DataFrame
articles = df['Article'].tolist()

# Clean articles
cleaned_articles = [clean_article(article) for article in articles]

# Check mood
mood_ratings = [check_mood(article) for article in articles]

# Find connections
topics = find_connections(articles)

# Visualize mood ratings as a donut chart
mood_count = pd.Series(mood_ratings).value_counts()
plt.figure(figsize=(8, 8))
plt.pie(mood_count, labels=mood_count.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title('Mood Ratings')
plt.axis('equal')
plt.show()

# Visualize topics
for topic_idx, topic in topics:
    words = re.findall(r'"([^"]*)"', topic)
    plt.figure(figsize=(8, 6))
    plt.barh(range(len(words)), [1]*len(words), align='center')
    plt.yticks(range(len(words)), words)
    plt.xlabel('Word Importance')
    plt.title('Topic {}'.format(topic_idx+1))
    plt.show()

# Aspect analysis (optional)
aspect_sentiments = aspect_analysis(articles)
print("Aspect Analysis:")
print(aspect_sentiments)

# Save aspect analysis results to CSV
aspect_sentiments.to_csv('aspect_analysis_results.csv', index=False)

# Example scoring logic (optional)
# You can implement additional scoring metrics as needed

# Check the system: Check if the summaries, mood ratings, and big ideas make sense
# Example scoring metrics: Precision, Recall, F1-score

# Visualize aspect sentiments (optional)
plt.figure(figsize=(10, 6))
aspect_count = aspect_sentiments['Aspect'].value_counts()
aspect_count.plot(kind='bar', color='skyblue')
plt.title('Aspect Sentiments')
plt.xlabel('Aspect')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
